```python
# Import Libararies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import dataset

ds=pd.read_csv('startup_data.csv')
ds
```

```python
#Divide dataset as independent(input) and dependent(output)

x=ds[['R&D Spend','Administration','Marketing Spend','State']]
y=ds[['Profit']]
```

```python
x
```

```python
y
```

```python
#Finding Unique values in state column

ds['State'].unique()
```

```python
#Encoding on column state

from sklearn .preprocessing import OneHotEncoder # OneHot Encoding is alphabeticaly
ohe=OneHotEncoder(sparse_output = False)
encoded_data = ohe.fit_transform(x[['State']])

column_names = ohe.get_feature_names_out(['State'])
encoded_df = pd.DataFrame(encoded_data,columns=column_names)

x=pd.concat([x,encoded_df],axis = 1)

x=x.drop(columns =['State'])
x
```

```python
#spliting dataset as Training and Testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
y_test
```

```python
#Introduce Model
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
#predict

data = [[100000,100000,100000,1,0,0]]
columns = x.columns
myinput = pd.DataFrame(columns =columns, data = data )
result = reg.predict(myinput)
result
```

```python
# For compare with x_test

y_pred = reg.predict(x_test)
y_pred
```

```python
y_test
```

```python
# For Accuration 

from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
score
```
